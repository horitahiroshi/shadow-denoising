# Shadow denoising

```python
# importing necessary libraries
import numpy as np
import imageio
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
I = imageio.imread("../images/shadow5.jpg")
plt.figure(figsize=(16,16))
plt.subplot(221)
plt.imshow(I)
plt.title("Shadowed image")
plt.subplot(222)
plt.imshow(I[:,:,0])
plt.title("R")
plt.subplot(223)
plt.imshow(I[:,:,1])
plt.title("G")
plt.subplot(224)
plt.imshow(I[:,:,2])
plt.title("B")
```

```python
def range_adjustment(img, minvalue=0, maxvalue=255):
    currmin = np.min(img)
    currmax = np.max(img)
    
    img_u8 = np.zeros(img.shape, dtype=np.uint8)
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            img_u8[x,y] = int(((img[x,y]-currmin)*(maxvalue - minvalue)/(currmax - currmin)) + minvalue)
    
    return img_u8    
```

```python
I_hsv = mpl.colors.rgb_to_hsv(I)
I_hsv = range_adjustment(I_hsv,0,255)
plt.figure(figsize=(16,16))
plt.subplot(221)
plt.imshow(I_hsv)
plt.title("Shadowed image")
plt.subplot(222)
plt.imshow(I_hsv[:,:,0])
plt.title("H")
plt.subplot(223)
plt.imshow(I_hsv[:,:,1])
plt.title("S")
plt.subplot(224)
plt.imshow(I_hsv[:,:,2])
plt.title("V")
```
